```python
import pandas as pd
import bar_chart_race as bcr
import numpy as np
import matplotlib.animation as anim
import matplotlib.pyplot as plt
from IPython.display import Video
from datetime import datetime
from matplotlib.animation import FuncAnimation
import warnings
import matplotlib.font_manager as fm
import requests
import bar_chart_race as bcr

warnings.filterwarnings('ignore')

df = pd.read_csv("2009-2024.csv").set_index('Date')
df.drop(columns="Unnamed: 1",inplace=True)

df = df.iloc[14:]

#Change Month to Indoesia
bulan_indonesia = {
    'January': 'Januari',
    'February': 'Februari',
    'March': 'Maret',
    'April': 'April',
    'May': 'Mei',
    'June': 'Juni',
    'July': 'Juli',
    'August': 'Agustus',
    'September': 'September',
    'October': 'Oktober',
    'November': 'November',
    'December': 'Desember'
    }
    
# Convert index to datetime (YYYY-MM format)
df.index = pd.to_datetime(df.index, format='%Y-%m')
date_str = df.index.strftime('%B %Y')
months, years = zip(*[ds.split() for ds in date_str])
months_id = [bulan_indonesia[month] for month in months]
translated_date_str = [f'{month} {year}' for month, year in zip(months_id, years)]
df.index = translated_date_str


#Bar chart Race
bcr.bar_chart_race(
    df=df,  
    title={'label':'Mobile Vendor Market Share Indonesia\n2010-2023',
           'size': 30,
            "family":"serif",
           "color":"white",
            'loc': 'center',
            'pad': 60
          },
    period_label={
                  'x': .85, 
                  'y': .15 , 
                  'ha': 'right', 
                  'va': 'center', 
                  'size': 23, 
                  "family":'serif',
                  "verticalalignment":'bottom',
                  "horizontalalignment":'right'
                 },
    filename="mobile_vendor_market_share.mp4", #Ubah ke .gif, jika Anda ingin hasilnya menjadi gif
    n_bars=10, #Maksimal bar yang dimunculkan
    colors = ['#00E100', '#1425A2', '#FF6700', '#C0C0C0', '#00C0FF', '#A2B1C4', 
     '#F4CE14', '#000000', '#4D4D4D', '#E72929', '#EE0000', '#0090c4',
     '#CE2328', '#1E201E', '#faf7f7', '#E5E1DA', '#7BD3EA', '#FBF9F1',
              
     '#A0153E', '#B4A5A5', '#E5E1DA', '#C8102E', '#0059A0', '#C8C8C8',
     '#FBC02D', '#F8F8F8', '#7B1FA2', '#A8A8A8', '#9C27B0', '#F4C300',
     '#D32F2F', '#D17B5A', '#0D47A1', '#4CAF50', '#673AB7', '#00796B',
     '#000000', '#E53935', '#F44336', '#0288D1', '#03A9F4', '#8BC34A'],
  
    img_label_folder="image",
    tick_image_mode="trailing",
    orientation='h',
    bar_kwargs={
        'alpha': 0.99,
        'lw':0
    },
    steps_per_period=45,
    period_length=1500,
    fig_kwargs={
        
        'dpi':120,
        'facecolor': '#021526' ,
        'figsize': (19, 8)
        
    },
    fixed_max=False,
    
    bar_label_font={'size':17},
    tick_label_font={'size':17},
    bar_texttemplate="{x: ,.2f}%",
    tick_template="{x: ,.2f}%",
)
```
